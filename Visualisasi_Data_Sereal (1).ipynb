```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("cereal.csv")
data.head()
```

```python
data.info()
```

```python
data['calories'].unique()
```

```python
data['rating'].unique()
```

```python
data['protein'].unique()
```

```python
unique_kalori_name = data[data.calories>0]['name'].unique()
unique_kalori_name.sort()
unique_kalori_name
```

```python
kalori_name = []
for i in unique_kalori_name:
  kalori_name.append(data[data['name']==i].calories.sum())
```

```python
plt.figure(figsize=(50,50))
plt.bar(unique_kalori_name, kalori_name)
plt.title('Rata Kalori Sereal')
plt.show()
```

```python
plt.figure(figsize=(50,50))
plt.pie(kalori_name, autopct='%1.1f&&')
plt.legend(unique_kalori_name)
plt.title('Rata Kalori Sereal Pada Manufacture')
plt.show()
```

```python
unique_merk_rating =  data['name'].unique()
unique_merk_rating.sort()
unique_merk_rating
```

```python
jml_merk_rating = []
for i in unique_merk_rating:
  jml_merk_rating.append(data[data['name']==i].rating.sum())
```

```python
plt.figure(figsize=(50,50))
plt.bar(unique_merk_rating, jml_merk_rating)
plt.title('Beberapa Merk Yang Terdapat Pada Manufacturer')
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(jml_merk_rating, autopct='%1.1f%%')
plt.legend(unique_merk_rating)
plt.title('Rating Merk')
plt.show()
```

```python
unique_manufacture = data['mfr'].unique()
unique_manufacture.sort()
unique_manufacture
```

```python
jml_manufacture= []
for i in unique_manufacture:
  jml_manufacture.append(data[data['mfr']==i].rating.sum())
```

```python
plt.figure(figsize=(7,7))
plt.bar(unique_manufacture, jml_manufacture)
plt.title('Rating Sereal Di Manufacture')
plt.show()
```

```python

plt.figure(figsize=(7,7))
plt.pie(jml_manufacture, autopct='%1.1f%%')
plt.legend(unique_manufacture)
plt.title('Rating Sereal Di Manufacture')
plt.show()
```
